```python
import pandas as pd
import numpy as np
import datetime
```

## Importing Billboard Data

```python
# To handle a character that utf-8 encoding could not handle (eg. Index 22: 98¡)
billboard = pd.read_csv('data/billboard.csv',encoding= 'unicode_escape')
billboard.head(25)
```

```python
billboard.describe()
```

```python
billboard.dtypes[0:10]
```

## Extracting Required Columns from Untidy Billboard Data - Melting

```python
melteddf = pd.melt(billboard,id_vars=list(billboard.columns[:6]),
        value_vars=list(billboard.columns[7:]),var_name='week',value_name='rank')
```

```python
melteddf.head()
```

```python
melteddf.drop('genre',axis=1,inplace=True)
```

## Extracting Week Numbers and Dropping NaN data from Dataframe

```python
list(melteddf.week.unique())[:5]
```

```python
melteddf = melteddf.assign(week = lambda x: x['week'].str.extract('(\d+)'))
melteddf.head()
```

```python
melteddf.isna().sum()
```

```python
melteddf = melteddf.dropna()
melteddf.isna().sum()
```

```python
len(melteddf)
```

## Updating Data types to required types and format

```python
melteddf.dtypes
```

```python
melteddf['date.entered'] = pd.to_datetime(melteddf['date.entered'],format='%m/%d/%y')
```

```python
melteddf = melteddf.astype({"week":'int64', "rank":'int64'})
melteddf.dtypes
```

```python
melteddf.head()
```

## Extracting Weekly Dates from Date.Entered & Week Columns

```python
melteddf['date'] = melteddf['date.entered'] + ((melteddf['week']-1)*7).map(datetime.timedelta)
```

## Selecting required columns and sorting to obtain final Tidy Billboard Data

```python
melteddf = melteddf[['year','artist','time','track','date','week','rank']]
part1_df = melteddf.sort_values(by=['artist','track','week'])
```
