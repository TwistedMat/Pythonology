## Removing $, symbols and Crypto Abbreviations to pick only values 

```python
rawdf["Price"] = rawdf["Price"].str.replace("[$,]","",regex=True)
rawdf["Market Cap"] = rawdf["Market Cap"].str.replace("[$,]","",regex=True)
rawdf["Circulating Supply"]=rawdf["Circulating Supply"].str.replace(",","",regex=True).str.split().str[0]
```

```python
# rawdf["Total Supply"]=rawdf["Total Supply"].str.replace("No Data","",regex=False).str.replace(",","").str.split().str[0]
# rawdf["Max Supply"]=rawdf["Max Supply"].str.replace("No Data","",regex=False).str.replace(",","").str.split().str[0]
# rawdf["All Time High Value"] = rawdf["All Time High Value"].str.replace("No Data","",regex=False).str.replace("[$,]","")
# rawdf["All Time Low Value"] = rawdf["All Time Low Value"].str.replace("No Data","",regex=False).str.replace("[$,]","")

# Indexing done to pick up the value and not the Crypto Abbreviations. Always 1st elementwhen split by " "
rawdf["Total Supply"]=rawdf["Total Supply"].str.replace(",","").str.split().str[0]
rawdf["Max Supply"]=rawdf["Max Supply"].str.replace(",","").str.split().str[0]
rawdf["All Time High Value"] = rawdf["All Time High Value"].str.replace("[$,]","",regex=True)
rawdf["All Time Low Value"] = rawdf["All Time Low Value"].str.replace("[$,]","",regex=True)

# Replaces all No Data or No values with None in the entire df
rawdf.replace({'No Data':None,'No':None},value=None,method=None,inplace=True)
```

## Converting Date Columns into DateTime

```python
rawdf[['All Time High Date','All Time Low Date']] = rawdf[['All Time High Date','All Time Low Date']].apply(pd.to_datetime)
```

## Converting Columns into Numerical Datatypes

```python
# Total and Max Supply kept as float to handle NaNs
conv_dict = {'Price': float,
             'Market Cap': float,
             'Circulating Supply': 'int64',
             'Total Supply': float,
             'Max Supply': float,
             'All Time High Value': float,
             'All Time Low Value':float}
```

```python
rawdf = rawdf.astype(conv_dict)
```

```python
rawdf.dtypes
```

## Converting remaining Volume columns into Int values
   ### Max Supply left as float to match required datatypes by instructor

```python
rawdf['Total Supply'] = rawdf['Total Supply'].astype('Int64')
rawdf['Max Supply'] = rawdf['Max Supply'].apply(lambda x: None if math.isnan(x) else int(x))
```

## Picking up required columns for writing the csv with the cleaned dataframe 

```python
cleandf = rawdf[['Name', 'Price', 'Market Cap', 'Circulating Supply','Total Supply', 'Max Supply', 'All Time High Value',
       'All Time High Date', 'All Time Low Value', 'All Time Low Date']].copy()
cleandf.rename(columns={"All Time High Value":"All Time High Price","All Time Low Value":"All Time Low Price"},inplace=True)
cleandf.head()
```
